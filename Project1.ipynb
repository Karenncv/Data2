```python
# The libraries 

import pandas as pd 
import numpy as np 
import plotly.express as px 
import plotly.graph_objects as go # Graphs other than scatter and bar or customize 
```

```python
# Save the file path 
p1 = '/content/IED.xlsx'

#Read file 

ied= pd.read_excel(p1)

#Check 

ied.info()
```

```python
ied = ied.sort_values(by='Year')
ied.head()
```

```python
# Drop does not delete date
ied = ied.set_index('Year', drop = False)
ied.head()
```

```python
px.line(ied['Sp'],
        animation_frame = ied['Year'].astype(str),
        range_x =['1992','2021'],
        range_y =[-1000,6000])
```

```python
fig = go.Figure(
    layout = go.Layout(
        updatemenus = [dict(type= "buttons", direction = "right", x = 0.9, y = 1.16)],
        xaxis = dict(range = ['1992','2022'],
                     autorange = False, tickwidth = 2,
                     title_text = 'Years'),
        yaxis = dict(range = [-1000,6000],
                     autorange = False, 
                     title_text = ''),
                     title = 'Foreign direct investment',
        title_font_size = 30,
        title_x = 0.5
        )
)
```

```python
fig.show()
```

```python
# Add traces

init = 1

fig.add_trace(
    go.Scatter(x = ied.Year[:init],
               y = ied.Sp[:init],
               name = 'Sp',
               visible = True, 
               line = dict(color = 'pink')
               )
)
```

```python
frames = [
    go.Frame(
        data = [
            go.Scatter(x =ied.Year[:k], y = ied.Sp[:k])
        ]
    )
    for k in range(init,len(ied)+1)

]
```

```python
#Animation

fig.update(frames = frames)
```

```python
# Buttons 

fig.update_layout(
    updatemenus = [
        dict(
            buttons = list([
                dict(label = 'Play',
                     method = 'animate',
                     args = [None,{'frame':{'duration':100}}]
                     )
                 ]
            )
        )
    ]
)

# The result

fig.show()
```
